```python
# Load the uploaded files and preview their contents to understand the structure of the data.
import pandas as pd

# Load the files
file1_path = '/content/final_nutrition (2).csv'
file2_path = '/content/Butik kvittorader (1).csv'

# Reading the CSV files
file1 = pd.read_csv(file1_path)
file2 = pd.read_csv(file2_path)

# Displaying the first few rows of each file to understand their structure
file1_head = file1.head()
file2_head = file2.head()

file1_columns = file1.columns
file2_columns = file2.columns

file1_head, file2_head, file1_columns, file2_columns
```

```python
from difflib import SequenceMatcher

# Extract relevant columns for matching
file1_data = file1[['Artikelbenamning']].dropna()
file2_data = file2[['itemDesc']].dropna()

# Function to calculate similarity
def similarity_ratio(a, b):
    return SequenceMatcher(None, str(a), str(b)).ratio()

# Create a list to store matches
matches = []

# Iterate over both datasets and calculate similarity
for idx1, artikelbenamning in file1_data['Artikelbenamning'].items():
    for idx2, itemdesc in file2_data['itemDesc'].items():
        sim = similarity_ratio(artikelbenamning, itemdesc)
        if sim >= 0.6:  # Threshold of 60%
            matches.append((idx1, idx2, artikelbenamning, itemdesc, sim))

# Convert matches to a DataFrame for better analysis
matches_df = pd.DataFrame(matches, columns=['File1_Index', 'File2_Index', 'Artikelbenamning', 'itemDesc', 'Similarity'])

# Display the first few matches
matches_df.sort_values(by='Similarity', ascending=False).head()
```

```python
# Display the full list of matches
matches_df_sorted = matches_df.sort_values(by='Similarity', ascending=False)
matches_df_sorted
```

```python
# Save the matches to a CSV file for download
output_path = '/content/matched_items.csv'
matches_df_sorted.to_csv(output_path, index=False)

output_path
```
